### 2.1. Data Acquisition

We obtained our data from a dataset on Kaggle (4). This dataset contains 41,099 songs from various genres and has 20 columns about the songs' attributes. Columns include data about track name, artist name, and more. Songs in the dataset date back as far as 1960 and go up to 2010. As this data was created by people on Kaggle, no web scraping was required to obtain it. For convenience, we stored the dataset on GitHub here: https://raw.githubusercontent.com/edshen17/FP4/master/genre_music.csv.


Below are the column names and descriptions. Please note that most of these attributes are directly from the Spotify API (5).
*   **track** - track name
*   **artist** -  artist name
*   **danceability** - how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable.
*   **energy** - measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy.
*   **key** - estimated overall key of the track. Integers map to pitches using standard Pitch Class notation.
*   **loudness** - loudness of a track in decibels (dB). Values range between -60 and 0 db.
*   **mode** - modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0.
*   **speechiness** - detection of the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value.
*   **acousticness** - confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic.
*   **instrumentalness** - prediction of whether a track contains no vocals.  The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.
*   **liveness** - detection of the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live.
*   **valence** - measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).
*   **tempo** - overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration.
*   **duration in seconds** - duration of the track in seconds.
*   **time signature** - estimated overall time signature of a track. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure).
*   **decade** - decade the song was created in.
*   **genre** - the genre/category a song belongs to.
*   **chorus_hit** - likeliness of chorus being a hit (%)
*   **sections** - how many sections the song has
*   **popularity** - number of times the song has been on a popularity list

For this data, we are using classification models to classify the songs by genre. The target variable is the genre and the feature variables are the attributes of the songs provided by the Spotify API/our dataset. More specifically, the core feature variables we use are: danceability, energy, speechiness, tempo, duration_s (feature selection from section 3.1.4). 

This is a supervised learning classification problem. The size of our data allows us to split it into training and testing sets.

The algorithms we plan to use:
*   **K-Nearest Neighbor** - Compares inputted data points to the k data points that are "closest" to the sample and determines the class based on those points (most frequent class is chosen). We chose this algorithm because with the large amount of data we have, comparing an input to similar data points should provide an accurate and reasonable classification. Also, simplicity makes this algorithm a solid choice.
*   **Support Vector Machine (SVM)** - Looks to create "boundaries" that maximize the differences between the given classes and their data points. More specifically, it tries to find the hyperplane/decision surface that maximizes the margin between classes. We chose to use this algorithm because it is very effective in high dimensional spaces such as the one we are working with.

*   **Decision Tree** - Constructs a tree that asks yes/no questions regarding the feature values. We chose this algorithm because it is fairly simple, effective, and could be visualized if necessary. The disadvantage, however, is that it may not be able to understand some more complex concepts, which we will keep in mind when exploring this algorithm.

### &emsp;3.1.0 Data Cleaning
Although our dataset comes from Kaggle, there are some columns that are unneeded for our analysis (eg. song popularity, artists, sections, etc.).
```python
#drop unneeded columns, NA columns, and duplicates (cleaning). We also rename the "genre" column to "target"
df = df_original.drop(columns=['track', 'artist', 'chorus_hit', 'sections', 'popularity']).dropna().drop_duplicates().rename(columns={'genre': 'target'})
df.head()
```

### &emsp;3.1.1 Data Wrangling

There are also some columns that need to be reformated. For example, the column "decade" gives us the decade the song was made in, but it does so in a string format (eg. "60s"). However, this column may be more valuable for us if we change it to contain int values. Similarly, the target variable (genre) contains string variables and hence cannot be used by scikitlearn. We would need to transform this categorical variable into a quantitative one so our ML algorithm can understand it.

```python
#function used to format the decade column (data wrangling)
def format_decade(decade):
    return int(decade[0:2])
df['decade'] = df['decade'].apply(format_decade)

#create dictionaries to hold values of the genres
genre_dict = {1: "pop", 2: "r&b", 3: "rock", 4: "latin", 5: "rap", 6: "edm"}
genre_nominal = {"pop": 1, "r&b": 2, "rock": 3, "latin": 4, "rap": 5, "edm": 6}

#to transform the genres into numbers
def transform_genre_name(column):
    return genre_nominal[column]

#create variables to store feature and target columns
df["target"] = df["target"].apply(transform_genre_name)
target = df['target']
features = df.loc[:, 'danceability':'decade']
df.head()
```

### &emsp;&emsp; 3.1.2 Preprocessing
Before running our models, our data should be preprocessed to increase performance. In this case, we would be modifying all of our numerical features to be on the same scale (values between 0 and 1). This is for two main reasons:
<br> 1) Our features have significantly different ranges. This can be problematic because it can lead to several features dominating the algorithm.
<br> 2) Our features are not measured in the same units, so our model may be unit-dependent and thus reliant on how the feature is measured.

In some ways, preprocessing greatly reduces the risk of the above problems. We split the dataset and performed preprocessing below. 

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

#split the dataset
def split_the_dataset():
    return train_test_split(features, target, random_state=3000)

X_train, X_test, y_train, y_test = split_the_dataset() #store the variables after splitting

#create the scaler, fit it with the training data, and scale the feature variables
scaler = MinMaxScaler()
scaler.fit(X_train) 
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### &emsp;&emsp; 3.1.3 Feature Extraction

One-hot encoding is used when the machine learning algorithm needs to determine the difference between categorical variables. This entails creating new features from existing features and using 0 or 1 to represent if the data belongs/does not belong to a certain class.  

The only string variable we have is the 'target,' but because that is the target variable, one-hot encoding is not applicable to our dataset. Thus, we could not apply any feature extraction techniques. 


### &emsp;&emsp; 3.1.4 Feature Selection

Because not all features may be equally important in predicting the music genres, we use feature extraction to get the most important ones. Doing so can increase accuracy and algorithm speed for large datasets. We have chosen to use Recursive Feature Elimination (RFE), which starts with all the features, builds the model, eliminates the least important feature, and repeats the process. Because of its recursive nature, RFE is computationally expensive compared to other feature elimination techniques. Nevertheless, RFE often leads to better results compared to other feature elimination techniques because of this.

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

#use rfe and the decision tree regressor to decide whether or not to keep the feature or not. We chose to only use the 5 most important features
select = RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select=5)
select.fit(X_train_scaled, y_train)

#scale only the selected features based on previous preprocessing
X_train_selected = select.transform(X_train_scaled)
X_test_selected = select.transform(X_test_scaled)

#print results of RFE
print('Selected features after RFE:')
bool_list = select.get_support()
for i in range(len(bool_list)):
    if (bool_list[i]): #if "True":
        print('\t' + features.columns[i])
```

```python
#function to perform statistical analysis
def t_test(data, IV1, IV2):
    from scipy import stats

    #run Welch's t-test
    t_stat, p_val = stats.ttest_ind(df[IV1], df[IV2], equal_var=False)

    #run a pearson test to get the pearson coefficient
    r, p = stats.pearsonr(df[IV1], df[IV2])

    #get degrees of freedom
    deg_free = len(df[IV1]) + len(df[IV2]) - 2

    #print out the stats
    print("t(%d)=%.2f, p=%.3f" % (deg_free, t_stat, p_val))
    print("Pearson coefficient:", str(r) + ",", "p=", p)
```

### 3.5. Model Optimization

We next chose to tune each of our models in order to determine if we can improve their accuracy. Additionally, our 100% accuracy on the training set for the decision tree model indicates that it is very likely to be overfitting the data. Tuning can be used to prevent this overfitting and improve the model's ability to generalize. We use Scikit-learn's GridSearchCV method to test various parameters and determine the best ones for each algorithm. 

Note: As we couldn't tell beforehand what hypertuning would do to our decision tree algorithm, we decided to tune all the algorithms just in case.

```python
from sklearn.model_selection import GridSearchCV
```

### &emsp;&emsp; 3.5.1. K-Nearest Neighbor

For the k-nearest neighbors algorithm, we can optimize the n_neighbors and metric parameters. As we recognize that lower n_neighbors values have the potential to overfit the data, we tried out a variety of values. We also tried 3 different distance functions (metric) to see if changing them would lead to better results.

```python
#create dictionary with all the possible tuning values
knn_param_grid = {"n_neighbors":[3, 5, 7, 9], "metric": ["euclidean", "manhattan", "minkowski"]}

grid_search_knn = GridSearchCV(KNeighborsClassifier(), knn_param_grid)

grid_search_knn.fit(X=X_train_selected, y=y_train)

# get the best values
best_n_neighbors = grid_search_knn.best_params_["n_neighbors"]
best_k_metric = grid_search_knn.best_params_["metric"]

print("K-Nearest Neighbor Classifier")
print("Best parameters: ", grid_search_knn.best_params_)
print("Best cross-validation score: ", grid_search_knn.best_score_)

print("Test set score: ", grid_search_knn.score(X_test_selected, y_test))
```

The results seem to indicate that the ideal value for n_neighbors is 9, which yields a slight increase in the cross-valdation and test set scores.

```python
for classifier in estimators: #iterate through our estimator object and use the classifiers
        estimators[classifier].fit(X=X_train_selected, y=y_train) #fit the model to the data/selected features
        accuracy = estimators[classifier].score(X_test_selected, y_test) #get accuracy of our model on test set
        print(classifier)
        print('\tPrediction accuracy on the test data: ' + format(accuracy*100, '.2f') + '%\n')
```
